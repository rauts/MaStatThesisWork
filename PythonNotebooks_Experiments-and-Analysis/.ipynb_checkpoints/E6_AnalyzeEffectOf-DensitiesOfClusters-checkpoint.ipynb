### E.6 Analyse the effect of "Densities of Clusters" in the datasets

#### Create Datasets

```python
###################################################################
# Dataset properties:
# 1. three clusters with equal number of samples in each cluster
# 2. two features that are normally distributed and uncorrelated
# 3. non-overlap between clusters but different variances 
###################################################################

n_samples = n_point * 3
overlap = 1.3

X, y = datasets.make_blobs(n_samples=n_samples,
                           n_features=2,
                           centers= np.array([[-2, -1], [1,0], [0,2]])*overlap,
                           cluster_std= np.array([0.8, 0.5, 0.3]),
                           random_state=seed_num)


D16_1 = X, y


###################################################################
# Dataset properties:
# 1. three clusters with equal number of samples in each cluster
# 2. two features that are normally distributed and uncorrelated
# 3. slight-overlap between clusters but different variances 
###################################################################
n_samples = n_point * 3
overlap = 0.95

X, y = datasets.make_blobs(n_samples=n_samples,
                           n_features=2,
                           centers= np.array([[-2, -1], [1,0], [0,2]])*overlap,
                           cluster_std= np.array([0.8, 0.5, 0.3]),
                           random_state=seed_num)


D16_2 = X, y



###################################################################
# Dataset properties:
# 1. four clusters with equal number of samples in each cluster
# 2. two features that are normally distributed and uncorrelated
# 3. non-overlap between clusters but different variances 

###################################################################

n_samples = n_point * 4
overlap = 1.4
X, y = datasets.make_blobs(n_samples=n_samples,
                           n_features=2,
                           centers= np.array([[-4, 2], [2,2], [-3, 4], [-4, 1]])*overlap,
                           cluster_std= np.array([4, 0.9, 0.7, 1]),
                           random_state=seed_num)

D17_1 = X, y


###################################################################
# Dataset properties:
# 1. 3 clusters with equal number of samples in each cluster
# 2. two features that are normally distributed and uncorrelated
# 3. non-overlap between clusters but different variances 

###################################################################

n_samples = n_point * 3
X, y = datasets.make_blobs(n_samples=n_samples,
                           n_features=2,
                           centers= np.array([[2,2], [-3, 3.5], [-4, 1]]),
                           cluster_std= np.array([1, 0.5, 0.7]),
                           random_state=seed_num)

D17_2 = X, y

```

Dataset Properties: 
    
    1 Clusters with varying densities, 
    2 Number of features = 2,
    3 Features of each cluster are normally distributed and i.i.d,
    4 Number of clusters is fixed equals to 3,  
    5 Equal number of samples in each cluster,
    6 ...
    7 Globular shape of clusters
    8 ...

Candidate algorithms:
    - Kmean, 
    - Agglomerative single-linkage,
    - Agglomerative complete-linkage,
    - Agglomerative average-linkage, 
    - Agglomerative ward-linkage,
    - DBSCAN,
    - Spectral clustering,
    - GMM
    
Observations:

`Dataset D16_1: well-separated and different cluster densities`

    All considered algorithm were abele to detect the clusters


`Dataset D16_2: slight overlaping clusters and different cluster densities`

    - Kmean: was able to detect cluster structure. 
    - Agglomerative single-linkage: was unable to detect cluster structure.
    - Agglomerative complete-linkage: was unable to detect cluster structure (3 clusters but high variance cluster was split).
    - Agglomerative average-linkage: was able to detect cluster but few points were misclassified in overlaping region.
    - Agglomerative ward-linkage: was able to detect clusters.
    - DBSCAN: was unable to detect clusters. However, manually tuning the parameters improved the results with classifing certain boundry points as noise points. 
    - Spectral clustering: was able to detect clusters.
    - GMM: was able to detect clusters.
    
    
`Dataset D17_1: high densities cluster overlay on a low density cluster`

    None of the algorithms were able to detect true clusters. However, Kmean, complete-linkage, ward-linkage, GMM and Spectral clustering were able to detect some high densities clusters. 
    
Conclusions:

    When varing density clusters are well-separated, all the candidate algorithms were able to detect the clusters. However, when the clusters are close to each other, only spectral clustering, GMM, Kmean, average-linkage and ward-linkage were able to detect the clusters. Moreover, the result of DBSCAN improved when the parameters were tuned manaually. When high dinsity clusters were overlay to a low density cluster, none of the candidate algorithms were able to detect the clusters. Nevertheless, Kmean, complete-linkage, ward-linkage, GMM and Spectral clustering were able to detect some high densities clusters.

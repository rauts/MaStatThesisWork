### E.7 Analyze the effect of "shape of clusters'' in datasets

#### Create Datasets

```python
# ==================================================================================
# Dataset properties:
# 1. four clusters with equal number of samples in each cluster
# 2. two features that are uniformally distributed
# 3. square shape clusters 
# ==================================================================================


D11 = create_square_cluster([n_point, n_point, n_point, n_point], reset_seed=True)


# ==================================================================================
# Dataset properties:
# 1. two clusters with equal number of samples in each cluster
# 2. two features that are not normally distributed
# 3. two rings are very close to each other 

# ==================================================================================
n_samples = n_point * 2 # two clusters

X, y = datasets.make_circles(n_samples=n_samples,
                             factor=.5,
                             noise=0.04,
                             random_state = seed_num)

D12 = X, y

# ==================================================================================
# Dataset properties:
# 1. two clusters with equal number of samples in each cluster
# 2. two features that are not normally distributed
# 3. two rings are far apart 

# ==================================================================================
n_samples = n_point * 2
X, y = datasets.make_circles(n_samples=n_samples, 
                             factor=.5,
                             noise=0.08,
                            random_state = seed_num)

D13 = X, y

# ==================================================================================
# Dataset properties:
# 1. two moon shaped clusters with equal number of samples in each cluster
# 2. two features that are not normally distributed
# 3. two moons are far apart 
# ==================================================================================

n_samples = n_point * 2
X, y = datasets.make_moons(n_samples=n_samples,
                           noise=.05,
                          random_state = seed_num)

D14 = X, y

# ==================================================================================
# Dataset properties:
# 1. two moon-shaped clusters with equal number of samples in each cluster
# 2. two features that are not normally distributed
# 3. two moons are very close together  

# ==================================================================================

n_samples = n_point * 2
X, y = datasets.make_moons(n_samples=n_samples, 
                           noise=0.1,
                          random_state=seed_num)

D15 = X, y

```

`Dataset D11: Square Clusters`

It can be seen that the clustering result of DBSCAN (based on optimized prarmeters) does not seem very well. Let us mannualy tune the parameters of DBSCAN algorithm. Two stages are followed to tune the parameters of DBSCAN. First, fixed the value of minpnt and vary the value of eps. Then, getting the relevant value of eps, vary the n_neighbour by keeping eps at fixed value obtained in the previous stage.

```python
# result of dbscan based on optimized parameters
test_para_D11, _ = get_parameter_dbscan(D11)
print_parameters(test_para_D11)
silhouette_plot(D11[0], get_cluster_label_dbscan(D11[0], test_para_D11))
```

```python
test_param = test_para_D11
test_param['min_samples'] = 3
test_dbscan_varying_eps(D11[0], (0.3,0.35), test_param)
```

Around `eps = 0.313`, the result seems to be improved. Let us fixed `eps` at this value and vary `n_neighbour`

```python
test_param = test_para_D11
test_param['eps'] = 0.313
test_dbscan_varying_minPnt(D11[0], (2,10), test_param)
```

The parameters `eps=0.313` and `minpnt=4` have slightely improved the results visually (silhouette score might be slightly lower for visually improved cluster results compared to the optimized parameters values that were based on Silhouette score). This indicate that Silhouette score based parameter tuning might not produce good results when the clusters are uniformaly distibuted.

`Dataset D12: Circular Clusters`

The result of DBSCAN and Spectral clustering are not satisfactory, so we further investigate these two algorthms one at a time, starting with DBSCAN.

`DBSCAN: Circular Clusters`

```python
para_dbscan_D12, _ = get_parameter_dbscan(D12)
print_parameters(para_dbscan_D12)
silhouette_plot(D12[0], get_cluster_label_dbscan(D12[0], para_dbscan_D12))
```

Note that the resulting cluster result does not resemble with the original clusters structure. 

```python
test_dbscan_varying_eps(D12[0], (0.1, 1), para_dbscan_D12)
```

The score is the same for the range of values of `eps` = `(0.194, 0.289]` and `minpnt = 2` that gives visually good results. However, `eps` > these range  the algorithm cerates a single cluster; all points belong to one cluster. This definitely raises the question whether to use silhouette score based performance of DBSCAN when the clusters are circular shaped. Let us see what happend when `eps` is fixed at 0.2 and `minpnt` is varied. 

```python
para_dbscan_D12['eps'] = 0.2
test_dbscan_varying_minPnt(D12[0],(2, 10), para_dbscan_D12)
```

It can be noticed that the range of `minpnt` = (2,5] and fixed value of `eps` = 0.2 were able to detect the cluster structure. However, further increase of `minpnt` the performance of the algorithm starts degrading: breaking the outer cluster into several clusters and increasing noise points.

`Spectral clustering: Circular Clusters`
    
It can be seen that the spectral clustering algorithm is unable to detect the cluster structures when the parameters were tuned based on Silhouette score. Let us tune these parameters, fix `n_component = 2` to true value and vary n_neighbours.

```python
para_sc_D12, _ = get_parameter_sc(D12, 2) 
print(para_sc_D12)
silhouette_plot(D12[0], get_cluster_label_sc(D12[0], para_sc_D12))
```

```python
para_sc_D12['n_clusters'] = 2
test_sc_varying_neighbour(D12[0], para_sc_D12, (2, 20))
```

The above figures illustrate that that giving the true value of `n_components = 2` and varying another parameter `n_neighbours`, the algorithm was able to find the true cluster for the range of values of `n_neighbour` = (5, 15). Interestingly, the silhouette score is higher `0.38` when optimized parameters were used than `0.11` when parameters were manually tuned. This highlights the fact that silhouette score based parameter tuning might not be apropriate when clusters are circular shape.  

`Dataset D14: Half Moon-Shaped Clusters`

Similar to the Dataset D12, Spectral clustering were unable to detect the cluster structure. Let us further investigate the algorithm.

`Spectral Clustering: Half Moon-Shaped Clusters`

```python
para_sc_D14, _ = get_parameter_sc(D14, 2) 
print(para_sc_D14)
silhouette_plot(D14[0], get_cluster_label_sc(D14[0], para_sc_D14))
```

```python
para_sc_D14['n_clusters'] = 2
test_sc_varying_neighbour(D14[0], para_sc_D14, (2, 15))
```

It can be seen that the algorithm was able to detect the cluster structure for `n_components = 2` and `n_neighbour = (7, 15)`. Although the silhouette score is less,`0.11`, when the parameter `n_neighbour` is manually tuned than, `0.38`, when the parameters were optimized based on silhouette score, the performance of the algorithm is found to be visually well. This means the parameter tuning based on silhouette score is not optimal when the clusters are circular shapes.

Observations:

`Dataset D11: Square Clusters`

    
    When the shape of in each cluster is square, and clusters are slightely close to each other:
    
    - Kmean: was able to detect square clusters. 
    - Agglomerative single-linkage: was unable to detect square clusters (only three clusters were detected).
    - Agglomerative complete-linkage: was able to detect clusters with few misclassification points.
    - Agglomerative average-linkage: was able to detect square clusters. 
    - Agglomerative ward-linkage: was able to detect square clusters.
    - DBSCAN: was able to detect clusters. By manually tuning parameters, silhouette score was higher.
    - Spectral clustering: was able to detect square clusters. 
    - GMM: was able to found square clusters.
    
    Tuning the parameters of DBSCAN manually (visually) improved the clustering result slightely. 
    
    
`Dataset D12: Circular Clusters`
    
    When the shape of in each cluster is circular:
    
    - Kmean: was unable to detect circular clusters, broke the clusters with a linear boundary (balanced split). 
    - Agglomerative single-linkage: was able to detect circular clusters.
    - Agglomerative complete-linkage: was unable to detect circular clusters (formed imbalanced clusters).
    - Agglomerative average-linkage: was unable to detect circular clusters (formed imbalanced clusters). 
    - Agglomerative ward-linkage: was unable to detect circular clusters (formed imbalanced clusters).
    - DBSCAN: was unable to detect circular clusters. Split clusters into many clusters. By manually tuning parameters, silhouette score was higher and algorithm was able to detect the clusters.
    - Spectral clustering: was unable to detect circular clusters. However, manually tuning the parameter `n_neighbours` the algorithm was able to detect the circular clusters. 
    - GMM: was unable to found circular clusters. Balanced clusters were formed. 
    
    Tuning the parameters of DBSCAN and Spectral clustreing manually (visually) improved the clustering results. 
    
    
`Dataset D14: Half-moon Clusters`
    
    When the shape of in each cluster is half-moon:
    
    - Kmean: was unable to detect circular clusters, broke the clusters with a linear boundary (balanced split). 
    - Agglomerative single-linkage: was able to detect half-moon clusters.
    - Agglomerative complete-linkage: was unable to detect circular clusters (formed imbalanced clusters).
    - Agglomerative average-linkage: was unable to detect circular clusters (formed imbalanced clusters). 
    - Agglomerative ward-linkage: was unable to detect circular clusters (formed imbalanced clusters).
    - DBSCAN: was able to detect half-moon clusters. 
    - Spectral clustering: was unable to detect half-moon clusters. However, manually tuning the parameter `n_neighbours` the algorithm was able to detect the half-moon clusters. 
    - GMM: was unable to found half-moon clusters. Balanced clusters were formed. 
    
    Tuning the parameters of Spectral clustreing manually (visually) improved the clustering results.
    
    
<font color=red>
    Conclusions:
   </font>

    When clusters are of non-globular shaped, the spectral clustering, DBSCAN, and single-linkage tend to perform well compared to other candidate clustering algorithms such as Kmean, complete-linkage, average-linkage, ward-linkage, and GMM. Particualrly, the results of DBSCAN and spectral clustering are sensitive to the value of the parameters. When the parameters of these two algorithm were tuned based on silhouette score, the clustering results were not as good as when these parameters were tuned manually. This indicates that the sihouette score based automatic parameter tuning is not appropriate for non-globular shaped clusters.

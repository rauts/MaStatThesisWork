### E.1 Analyze the effect of "number of clusters'' in datasets

#### Create Datasets

```python
# ==================================================================================
# Dataset properties:
# 1. two clusters with equal number of samples in each cluster
# 2. two features that are normally distributed and uncorrelated
# 3. no overlap between clusters 
# ==================================================================================
n_clusters = 2
overlap = 1

dataset_para_D1 = {  'cluster_sample_sizes':np.array([n_point]*n_clusters),
                     'cluster_nums':n_clusters,
                     'cluster_means': np.array([[-4, 3], [4,-5]])*overlap,
                     'cluster_stds': np.array([[1, 1], [1,1]
                                              ]),
                      'cluster_xy_scalings': np.array([(1, 1), (1, 1)]),
                      'cluster_rotation_angles': np.array([0, 0]), # angles in radians
                      'cluster_seed_numbers': seed_num,
                      'cluster_overlaps': overlap,

                     }


D1 = create_multiclass_bivariate_data(dataset_para_D1, reset_seed=True)

# ==================================================================================
# Dataset properties:
# 1. three clusters with equal number of samples in each cluster
# 2. two features that are normally distributed and uncorrelated
# 3. no overlap between clusters 
# ==================================================================================
n_clusters = 3
overlap = 1

dataset_para_D2 = {  'cluster_sample_sizes':np.array([n_point]*n_clusters),
                     'cluster_nums':n_clusters,
                     'cluster_means': np.array([[-10, -4], [-4, -8], [0, 0]])*overlap,
                     'cluster_stds': np.array([[1, 1]]*n_clusters),
                     'cluster_xy_scalings': np.array([[1, 1]]*n_clusters),
                     'cluster_rotation_angles': np.array([0]*n_clusters), # angles in radians
                     'cluster_seed_numbers': seed_num,
                     'cluster_overlaps': overlap
                     }


D2 = create_multiclass_bivariate_data(dataset_para_D2, reset_seed=True)


# ==================================================================================
# Dataset properties:
# 1. five clusters with equal number of samples in each cluster
# 2. two features that are normally distributed and uncorrelated
# 3. no overlap between clusters 
# ==================================================================================
n_clusters = 5
overlap = 1

dataset_para_D3 = {  'cluster_sample_sizes':np.array([n_point]*n_clusters),
                     'cluster_nums':n_clusters,
                     'cluster_means': np.array([[-10, -10], [10, 10], [5, -5], [-7, 0], [2,8]])*overlap,
                     'cluster_stds': np.array([[1, 1]]*n_clusters),
                      'cluster_xy_scalings': np.array([(1, 1)]*n_clusters),
                      'cluster_rotation_angles': np.array([0]*n_clusters), # angles in radians
                      'cluster_seed_numbers': seed_num,
                      'cluster_overlaps': overlap,

                     }


D3 = create_multiclass_bivariate_data(dataset_para_D3, reset_seed=True)
```

#### Visualize Datasets

```python
# Select and create datasets of varing number of clusters
dataset_num_cluster = [
                       (D1, {'name': 'D1', 'n_cluster': 2}), 
                       (D2, {'name': 'D2', 'n_cluster': 3}),  
                       (D3, {'name': 'D3', 'n_cluster': 5}),  
#                        (D4, {'name': 'D4', 'n_cluster': 8}),
#                        (D5, {'name': 'D5', 'n_cluster': 11}), 
#                        (D6, {'name': 'D6', 'n_cluster': 16})
                      ]
plot_datasets(dataset_num_cluster)
```

However, silhouette score was `0.5` when the parameter were tuned based on silhouette score and the performance was not good. This illustrates that a higher value of silhouette score does not necessarily a good performance indicator. Thus, care must be taken to interprete the results based on silhouette score.

Observations:

`Dataset D10_2: Highly correlated features or elongated clusters`
    - Kmean: was unable to detect clusters, split the clusters with a linear boundary 
    - Agglomerative single-linkage: was able to detect clusters,
    - Agglomerative complete-linkage: was unable to detect clusters, split the clusters
    - Agglomerative average-linkage: was able to detect clusters 
    - Agglomerative ward-linkage: was able to detect clusters
    - DBSCAN: was able to detect clusters
    - Spectral clustering: was able to detect clusters
    - GMM: was able to detect clusters
    
`Dataset D10_2: Elongated clusters`
    - Kmean: was unable to detect clusters, split the clusters with a linear boundary 
    - Agglomerative single-linkage: was able to detect clusters,
    - Agglomerative complete-linkage: was unable to detect clusters, split the clusters
    - Agglomerative average-linkage: was unable to detect clusters 
    - Agglomerative ward-linkage: was unable to detect clusters
    - DBSCAN: was unable to detect clusters
    - Spectral clustering: was able to detect clusters
    - GMM: was unable to detect clusters
    
Only `Spectral clustering` was able to detect the clusters. However, tuning the parameters of `DBSCAN` and increasing the initialization parameter of `GMM` both `DBSCAN` and `GMM` were able to detect the cluster structures.
   
`Dataset D10_4: Mix of spherical and elongated clusters`
    - Kmean: was able to detect clusters
    - Agglomerative single-linkage: was unable to detect clusters
    - Agglomerative complete-linkage: was able to detect clusters
    - Agglomerative average-linkage: was unable to detect clusters 
    - Agglomerative ward-linkage: was unable to detect clusters
    - DBSCAN: was able to detect clusters
    - Spectral clustering: was able to detect clusters
    - GMM: was able to detect clusters

Only `Single-linkage` was unable to detect the clusters.
    
Conclusions:
    When the normally distributed features are highly correlated or elongated clusters, Kmean did not perform well since it tried to break the clusters with linear boundaries. Single-linkage, DBSCAN, Spectral clustering and GMM yielded good results. Other agglomerative clustering algorthms did not perform well. Furthermore, we noticed that the higher value of sihouette score did not show good results when the parameters of DBSCAN were tuned based on higher value of silhouette score. This highlights the fact that care must be taken when silhouette based parameter tuning is applied.

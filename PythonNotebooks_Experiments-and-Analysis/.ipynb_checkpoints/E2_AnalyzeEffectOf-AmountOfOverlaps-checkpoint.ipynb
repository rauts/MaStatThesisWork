   
Dataset Properties: 
    
    1 Clusters with varying amount of overlaps, 
    2 Number of features = 2,
    3 Features of each cluster are normally distributed and i.i.d,
    4 Number of clusters = 3, fixed,  
    5 Equal number of samples in each cluster
    6 Equal variance in each cluster
    7 Globular shape of clusters
    8 ...

Observations:

`Dataset D2_1: No overlap`
    - Kmean: was able to detect non-overlaping clusters, 
    - Agglomerative single-linkage: was able to detect non-overlaping clusters, 
    - Agglomerative complete-linkage: was able to detect non-overlaping clusters, 
    - Agglomerative average-linkage: was able to detect non-overlaping clusters,  
    - Agglomerative ward-linkage: was able to detect non-overlaping clusters, 
    - DBSCAN: was able to detect non-overlaping clusters, 
    - Spectral clustering: was able to detect non-overlaping clusters, 
    - GMM: was able to detect non-overlaping clusters
    
`Dataset D2_2: slight overlap`
    - Kmean: was able to detect slightely overlaped clusters, but overlaping points were misclassified 
    - Agglomerative single-linkage: was unable to detect slightley overlaped clusters, 
    - Agglomerative complete-linkage: was able to detect slightely overlaped clusters, but overlaping points were misclassified 
    - Agglomerative average-linkage: was able to detect largely overlaped clusters, but overlaping points were misclassified  
    - Agglomerative ward-linkage: was able to detect slightely overlaped clusters, but overlaping points were misclassified
    - DBSCAN: was unable to detect slightely overlaped clusters, 
    - Spectral clustering: was able to detect slightely overlaped clusters, but overlaping points were misclassified  
    - GMM: was able to detect slightely overlaped clusters, but overlaping points were misclassified with certain probabilities.
    
`Dataset D2_3: large overlap`
    - Kmean: was able to detect largely overlaped clusters, but overlaping points were misclassified 
    - Agglomerative single-linkage: was unable to detect largely overlaped clusters, a large cluster and one point clusters. 
    - Agglomerative complete-linkage: was able to detect largely overlaped clusters, but overlaping points were misclassified 
    - Agglomerative average-linkage: was unable to detect largely overlaped clusters, created a single large cluster and few points in other clusters  
    - Agglomerative ward-linkage: was able to detect largely overlaped clusters, but overlaping points were misclassified
    - DBSCAN: was unable to detect largely overlaped clusters, formed a large cluster with noise points 
    - Spectral clustering: was able to detect largely overlaped clusters, but overlaping points were misclassified  
    - GMM: was able to detect largely overlaped clusters, but overlaping points were misclassified with certain probabilities.

    
Conclusions:
    
    All candidate algorithms were able to detect clusters in the dataset that have above properties and non-overlaping clusters. When clusters are very closed to each other or slightely overlaps (few boundaries points among clusters), except agglomerative single-linkage algorithm that formed unbalanced clusters (a single point in two clusters), all other considered clustering algorithms were able to detect clusters. However, the performance of these algorithms was decreased compared to earlier case (non-overlaping clusters). Particularly, the border points of clusters were misclassified by the algorithms. Likewise, further increase of overlap among clusters, the performance of the algorithms is further decreased.

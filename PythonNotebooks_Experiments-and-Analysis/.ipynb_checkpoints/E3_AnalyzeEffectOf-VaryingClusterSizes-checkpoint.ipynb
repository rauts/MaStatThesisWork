#### Create Datasets

```python
###################################################################
# Dataset properties:
# 1. two clusters with unequal number of samples in each cluster
# 2. two features that are normally distributed and uncorrelated
# 3. no overlap between clusters 
###################################################################

# create parameters of the dataset
n_clusters = 2
overlap = 1.1

dataset_para_D9_1 = {  'cluster_sample_sizes':np.array([n_point, 5*n_point]),
                     'cluster_nums':n_clusters,
                     'cluster_means': np.array([[0, 0], [5, 0]])*overlap,
                     'cluster_stds': np.array([[1, 1]]*n_clusters),
                     'cluster_xy_scalings': np.array([[1, 1]]*n_clusters),
                     'cluster_rotation_angles': np.array([0]*n_clusters), # angles in radians
                     'cluster_seed_numbers': seed_num,
                     'cluster_overlaps': overlap
                     }

# create dataset
D9_1 = create_multiclass_bivariate_data(dataset_para_D9_1, reset_seed=True)


# create parameters of the dataset
n_clusters = 2
overlap = 0.98

dataset_para_D9_1_1 = {  'cluster_sample_sizes':np.array([n_point, 5*n_point]),
                     'cluster_nums':n_clusters,
                     'cluster_means': np.array([[0, 0], [5, 0]])*overlap,
                     'cluster_stds': np.array([[1, 1]]*n_clusters),
                     'cluster_xy_scalings': np.array([[1, 1]]*n_clusters),
                     'cluster_rotation_angles': np.array([0]*n_clusters), # angles in radians
                     'cluster_seed_numbers': seed_num,
                     'cluster_overlaps': overlap
                     }

# create dataset
D9_1_1 = create_multiclass_bivariate_data(dataset_para_D9_1_1, reset_seed=True)

###################################################################
# Dataset properties:
# 1. three clusters with unequal number of samples in each cluster
# 2. two features that are normally distributed and uncorrelated
# 3. no overlap between clusters 
###################################################################

# create parameters of the dataset
n_clusters = 3
overlap = 1

dataset_para_D9_2 = {  'cluster_sample_sizes':np.array([n_point, 5*n_point, 7*n_point]),
                     'cluster_nums':n_clusters,
                     'cluster_means': np.array([[0, 0], [5, 0], [-3, -3]])*overlap,
                     'cluster_stds': np.array([[1, 1]]*n_clusters),
                     'cluster_xy_scalings': np.array([[1, 1]]*n_clusters),
                     'cluster_rotation_angles': np.array([0]*n_clusters), # angles in radians
                     'cluster_seed_numbers': seed_num,
                     'cluster_overlaps': overlap
                     }

# create dataset
D9_2 = create_multiclass_bivariate_data(dataset_para_D9_2, reset_seed=True)


###################################################################
# Dataset properties:
# 1. four clusters with unequal number of samples in each cluster
# 2. two features that are normally distributed and uncorrelated
# 3. no overlap between clusters 
###################################################################

# create parameters of the dataset
n_clusters = 4
overlap = 1

dataset_para_D9_3 = {  'cluster_sample_sizes':np.array([n_point//3, 4*n_point, 7*n_point, n_point]),
                     'cluster_nums':n_clusters,
                     'cluster_means': np.array([[0, 0], [5, 0], [-3, -3], [4, -4 ]])*overlap,
                     'cluster_stds': np.array([[1, 1]]*n_clusters),
                     'cluster_xy_scalings': np.array([[1, 1]]*n_clusters),
                     'cluster_rotation_angles': np.array([0]*n_clusters), # angles in radians
                     'cluster_seed_numbers': seed_num,
                     'cluster_overlaps': overlap
                     }

# create dataset
D9_3 = create_multiclass_bivariate_data(dataset_para_D9_3, reset_seed=True)
```

Observations:

`Dataset D9_1: two separated clusters with different sample sizes`

    - Kmean: was able to detect cluster structures, 
    - Agglomerative single-linkage: was unable to detect cluster structure, a single large cluster and one-point cluster
    - Agglomerative complete-linkage: was able to detect cluster structure
    - Agglomerative average-linkage: was able to detect cluster structure 
    - Agglomerative ward-linkage: was able to detect cluster structure
    - DBSCAN: was able to detect cluster structure
    - Spectral clustering: was able to detect cluster structure
    - GMM: was able to detect cluster structure

`Dataset D9_1_1: two close clusters with different sample sizes`

    - Kmean: was able to detect cluster structures, 
    - Agglomerative single-linkage: was unable to detect cluster structure, a single large cluster and one-point cluster
    - Agglomerative complete-linkage: was able to detect cluster structure, but some boundary points were misclassified
    - Agglomerative average-linkage: was able to detect cluster structure, but some boundary points were misclassified
    - Agglomerative ward-linkage: was able to detect cluster structure, but some boundary points were misclassified
    - DBSCAN: was unable to detect cluster structure
    - Spectral clustering: was able to detect cluster structure
    - GMM: was able to detect cluster structure

`Dataset D9_3: four clusters with different sample sizes`
    
    - Kmean: was unable to detect cluster structures, the cluster with large samples tends to break 
    - Agglomerative single-linkage: was unable to detect cluster structure, a single large cluster and one-point cluster
    - Agglomerative complete-linkage: was able to detect cluster structure, performed better than Kmean
    - Agglomerative average-linkage: was able to detect cluster structure, performed better than Kmean
    - Agglomerative ward-linkage: was able to detect cluster structure, performed better than Kmean
    - DBSCAN: was unable to detect cluster structure
    - Spectral clustering: was unable to detect cluster structure, cluster with large smaple size was broken
    - GMM: was able to detect cluster structure


    
Conclusions:
    When clusters are with varing sample sizes, Kmean, spectral clustering and gmm tend to break the large cluster. Single-linkage, however, was unable to detect the clusters, while complete, average and ward linkage methods were able to detect the clusters. It was observed that the spectral clustering algorithm is sensitive to the value of the parameter, `n_neighbours`. In addition, DBSCAN failed to capture the structure of clusters.

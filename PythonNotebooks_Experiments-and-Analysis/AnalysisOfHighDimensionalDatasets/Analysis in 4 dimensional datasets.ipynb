## Analyze effect of 4 dimensional datasets

Following steps are used to analyze the effect of the three dimensional data

    1. Create 3 dimensional datasets
        a) add 4th dimension by one of the repeated columns from the 3 dimensional dataset
        b) add 4th dimension by a linear combination of columns of the 3 dimensional dataset
        c) add 4th dimension by a gaussian noise vector with zero mean and unit variance
    2. Inspect the datasets visually
    3. Run the candidate clustering algorithms
    4. Analyze the results using silhoueet plot
    
To generate datasets, we use the make_blobs functions from sklearn module.

### Experiment 1: Analyze the effect of 4th dimension using 3 dimensional dataset

The 3 dimensional dataset will be the same for all three cases.

Properties of the datasets:

    1. Number of clusters = 3

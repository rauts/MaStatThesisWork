By experimenting with different values of `eps`, it can be shown that the value of `eps` around 0.505 and `minpnt=2` would give better results.

```python
test_para_dbscan_D20 = para_dbscan_D20
test_para_dbscan_D20['eps'] = 0.505
test_dbscan_varying_minPnt(D20[0],(2, 10), test_para_dbscan_D20)
```

It can be noticed that by fixing the `eps = 0.505` and varying the parameter `minpnt (number of neighbours in the radius eps)` from 2 to 10 the DBSCAN was able to remove the small size (2 or 3 point clusters) clusters, nevertheless, the algorithm was not able to further improve the result since those small size clusters were turned into noise points by the algorithm. Additionally, the consequence of increasing `minpnt` beyound 5 is that the points in the high variance cluster started turning into noise points while the points in low variance cluster were not affected.

Observations:

`Dataset D20: two clusters with different variances`

    - Kmean: was able to detect the clusters, however, the cluster with large variance was broken 
    - Agglomerative single-linkage: was unable to detect clusters, formed a single large cluster and one point cluster
    - Agglomerative complete-linkage: was able to found the clusters, but the large variance cluster was split
    - Agglomerative average-linkage: was unable to detect clusters, formed a single large cluster and one point cluster
    - Agglomerative ward-linkage: was able to detect clusters and performed better than other condidate agglomerative clustering algorithms
    - DBSCAN: was unable to detect the clusters, just formed a single cluster
    - Spectral clustering: was able to detect the cluster structures
    - GMM: was able to detect the cluster structures
    
By tuning the parameters of DBSCAN manually improved the result to a certain extent, howerver, we also noticed that there is a trade-off between the amount of noise points and required number of clusters when there are clusters with different variances.  
    
`Dataset D20_1: three clusters with different variances`

    - Kmean: was able to detect the clusters 
    - Agglomerative single-linkage: was unable to detect clusters, could not form balanced clusters
    - Agglomerative complete-linkage: was able to found the clusters
    - Agglomerative average-linkage: was able to detect clusters, but performed slightely less than complete-linkage
    - Agglomerative ward-linkage: was able to detect clusters
    - DBSCAN: was unable to detect the clusters. Detected cluster were not balanced in size
    - Spectral clustering: was able to detect the cluster structures
    - GMM: was able to detect the cluster structures
    

Conclusions:
    When the variance in each cluster varies, Kmean and complete linkage tend to break the cluster with larger variance, and single linkage and average were unable to detect the clusters while ward-linkage, spectral clustering and gmm tend to perform well in this situation. DBSCAN is unable to detect the clusters, however, manaually tuning the parameters the result was improved. Furthermore, we notice that when the clusters are well-separated despite having variable variances in clusters, all candidate algorithms were able to detect the clusters.

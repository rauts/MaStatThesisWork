```python

# ==================================================================================
# Dataset properties:
# 1. two clusters with equal number of samples in each cluster
# 2. two features that are not normally distributed
# 3. two rings are very close to each other 

# ==================================================================================
n_samples = n_point * 2 # two clusters

X, y = datasets.make_circles(n_samples=n_samples,
                             factor=.3,
                             noise=0.1,
                             random_state = seed_num)

D12 = X, y

# ==================================================================================
# Dataset properties:
# 1. two clusters with equal number of samples in each cluster
# 2. two features that are not normally distributed
# 3. two rings are far apart 

# ==================================================================================
n_samples = n_point * 2
X, y = datasets.make_circles(n_samples=n_samples, 
                             factor=.5,
                             noise=0.08,
                            random_state = seed_num)

D13 = X, y

# ==================================================================================
# Dataset properties:
# 1. two moon shaped clusters with equal number of samples in each cluster
# 2. two features that are not normally distributed
# 3. two moons are far apart 
# ==================================================================================

n_samples = n_point * 2
X, y = datasets.make_moons(n_samples=n_samples,
                           noise=.07,
                          random_state = seed_num)

D14 = X, y

# ==================================================================================
# Dataset properties:
# 1. two moon-shaped clusters with equal number of samples in each cluster
# 2. two features that are not normally distributed
# 3. two moons are very close together  

# ==================================================================================

n_samples = n_point * 2
X, y = datasets.make_moons(n_samples=n_samples, 
                           noise=0.1,
                          random_state=seed_num)

D15 = X, y

# ==================================================================================
# Dataset properties:
# 1. three clusters with equal number of samples in each cluster
# 2. two features that are normally distributed and uncorrelated
# 3. non-overlap between clusters 
# 4. 20% of total samples as uniformly distributed random noise
# ==================================================================================

n_samples = n_point * 3
noise_level = 0.2 # noise percentage
f_samples = int(n_samples*(1-noise_level))
n_noise = n_samples - f_samples

X, y = datasets.make_blobs(n_samples=f_samples,
                           n_features=2,
                           centers= np.array([[-2, -1], [1,0], [0,2]]),
                           cluster_std= np.array([0.3, 0.2, 0.1]),
                           random_state=seed_num)

# crate noisy sampels of uniformaly distributed random variables  
noise_feature1 = np.random.uniform(low = -3, high = 3.5,  
                            size = (n_noise, 1))
noise_feature2 = np.random.uniform(low = -3, high = 3.5,  
                            size = (n_noise, 1))

noise_samples = np.concatenate([noise_feature1, noise_feature2], axis=1)
# Include noisy instances into the dataset
X = np.concatenate([X, noise_samples], axis=0)
# noise are labeled as -1
y = np.concatenate([y, np.repeat(a=-1, repeats=len(noise_samples))],
                   axis=0)

D16 = X, y

# ==================================================================================
# Dataset properties:
# 1. four clusters with equal number of samples in each cluster
# 2. two features that are normally distributed and uncorrelated
# 3. non-overlap between clusters but different variances 

# ==================================================================================

n_samples = n_point * 4
X, y = datasets.make_blobs(n_samples=n_samples,
                           n_features=2,
                           centers= np.array([[-4, 2], [4,4], [-3, 4], [-4, 1]]),
                           cluster_std= np.array([2, 0.6, 0.2, 0.3]),
                           random_state=seed_num)

D17 = X, y

# ==================================================================================
# Dataset properties:
# 1. no particular cluster structure or single cluster

# ==================================================================================

n_samples = n_point * 2
X, y = np.random.rand(n_samples, 2), np.repeat(a=0, repeats=n_samples) 

D18 = X, y


# ==================================================================================
# Dataset properties: unisotropic clusters
# 1. three clusters with equal number of samples in each cluster
# 2. two features that are normally distributed
# 3. clusters are correlated, unisotropic clusters

# ==================================================================================
n_samples = n_point * 3

X, y = datasets.make_blobs(n_samples=n_samples, centers=3,  
                           random_state=seed_num)
transformation = [[0.6, -0.85], [-0.3, 0.6]]
X_aniso = np.dot(X, transformation)
X, y = X_aniso, y

D19 = X, y


# ==================================================================================
# Dataset properties: different variance
# 1. three clusters with equal number of samples in each cluster
# 2. two features that are normally distributed
# 3. variances of clusters are different
# ==================================================================================
n_samples = n_point * 3
cluster_mean = np.array([[0, 0], [3, 3], [5, 0]])
cluster_std = np.array([0.2, 1.5, 0.5]) 
      
X, y = datasets.make_blobs(n_samples=n_samples,
                           n_features=2,
                           centers=cluster_mean,
                           cluster_std=cluster_std, 
                           random_state=seed_num)
D20 = X, y
```

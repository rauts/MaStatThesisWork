```python
# ==================================================================================

def optimize_parameters_and_run_algorithms(data, default_data = {}, display_results=True):
    """ 
    Return y_pred (cluster labels) and Silhouette metrics for a dataset
    
    Parameters:
    -----------
    
    data : a list of tuple of two elements
    
        The first element is an array, data matrix, shape (n_samples, n_attributes)
        and the second element is a dictionary, consisting of the parameters of the datasets
        
    default_data: a dictionary
        It contains default parameters for the algorithms to be optimized
        
    display_results: a boolean
        If true it prints outputs of parameter optimization step.
        
        
    Returns:
    -------
    
    y_pred_results, metric_results: a tuple
    
        metric_results is a list of tuple consisting of the name of the algorithms
        and performance metrics 
        y_pred_results is a list of tuple consisting of the name of the algorithms and 
        and predicted labels obtained from the algorithms for each dataset.
        
    """

    # store results of clustering algorithms 
    metric_results = []
    y_pred_results = []

    # t0 = time.time()

    # Analyze the clustering algorithms on the dataset
    for i_dataset, (dataset, data_params) in enumerate(data):
        print('')
        print('Dataset {} ====='.format(data_params['name']))
        
        
        # update parameters with dataset-specific values
        params = default_data.copy()
        params.update(data_params)

        # unpack dataset
        X, y_true = dataset       
        
        # get the optimal parameters for individul algorithms
        t0 = time.time()
        opt_params = optimize_algorithms(dataset, params, display_results)
        t1 = time.time()
        # get the clustering labels and silhouette metrics based on optimized paramters
        y_pred, metrics = run_algorithms(dataset, opt_params) # km, dbscan, sc, gmm, sl, cl, al and wl in this order
        t2 = time.time()

        # store results of each datasets
        metric_results.append(metrics)
        y_pred_results.append(y_pred)

        # Execution time summary
        print()
        print('############################################################')
        print('#####    Parameter Optimization Summary: Dataset {}   #####'.format(data_params['name'])) 
        print('############################################################')
        # Optimal parameters
        
        print('')
        print('Parameters:=====')
        print('')
        for algo in opt_params:
            print(algo +': ######')
            print_parameters(opt_params[algo]) 
            print()
            
        if X.shape[1] == 2:
            print()
            print('Performance Metrices =====')
            print()
            print('     Silhouette l2, Silhouette l1, Silhouette cosine' )
            print('')
        
            for i, (algo, res) in enumerate(metrics):
                print('{:<15s}: {}'.format(algo, np.array(list(res))))

        # plot clustered results only in 2d 
        if X.shape[1] == 2:
            print()
            print('Clustered Data =====')
            print()
            plot_n_cluster_results(dataset, y_pred)  
              
    #results are in this order: km, dbscan, sc, gmm, sl, cl, al and wl  
    return y_pred_results, metric_results
```
